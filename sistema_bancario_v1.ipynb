*Código de Sistema Bancário - Versão 01* 
-  Nocões básicas de operadores, blocos de comando, strings e funções.

```python
print("""Bem vindo ao sistema bancário!
Por favor, escolha uma opção:
    [1] Sacar
    [2] Depositar
    [3] Visualizar extrato
    [0] Sair""")

saldo = 0
limite_saques = 3
n_saque = 0
limite = 500.00
extrato = ""

while True:
    
    input_usuario = int(input())

    def sacar(saque):
        global saldo, n_saque, limite_saques, extrato

        if saque <= saldo and saque > 0 and saque <= limite and n_saque < limite_saques:
            print(f"Saque de R$ {saque:.2f} realizado com sucesso!")
            saldo -= saque
            n_saque += 1
            print(f"Seu saldo atual é de R$ {saldo:.2f}")
            print(f"Você ainda tem {limite_saques - n_saque} saques disponíveis.")
            extrato += ((f"Saque de R$ {saque:.2f} realizado. Seu saldo atual é de R$ {saldo:.2f} \n"))
        elif saque <= 0:
            print("Valor de saque inválido!")
        elif saque > limite:
            print("Limite de saque excedido!")
        elif n_saque >= limite_saques:
            print("Número máximo de saques diários atingido!")
        else:
            print("Saldo insuficiente!")

    def depositar(deposito):
        global saldo, extrato
        if deposito > 0:
            saldo += deposito
            print(f"Depósito de R$ {deposito:.2f} realizado com sucesso!")
            extrato += ((f"Depósito de R$ {deposito:.2f} realizado. Seu saldo atual é de R$ {saldo:.2f} \n"))
            print(f"Seu saldo atual é de R$ {saldo:.2f}")
        else:
            print("Insira um valor válido para depósito.")

    def mostrar_extrato():
        global extrato
        if not extrato:
            print("=== Extrato ===")
            print("Não há registros de operações.")
            print("================")
        else:
            print("=== Extrato ===")
            print(extrato)
            print("================")

    if input_usuario == 1:
        print("Você escolheu a opção [1] saque.")
        saque = float(input("Digite o valor que deseja sacar: "))
        sacar(saque)
    elif input_usuario == 2:
        print("Você escolheu a opção [2] depósito.")
        deposito = float(input("Digite o valor que deseja depositar: "))
        depositar(deposito)
    elif input_usuario == 3:
        print("Você escolheu a opção [3] extrato.")
        mostrar_extrato()
        print(f"Seu saldo atual é de R$ {saldo:.2f}. Você pode realizar mais {limite_saques - n_saque} saques hoje. \n")
    elif input_usuario == 0:
        print("Você escolheu a opção [0] sair.")
        print("Obrigado por usar nosso sistema bancário. Até logo!")
        break
    else:
        print("Opção inválida. Por favor, escolha uma opção válida.")
```
